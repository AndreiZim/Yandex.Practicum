# Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## Цель исследования

На основании имеющихся данных о продажах игр, оценках пользователей и экспертов, жанрах и платформах, выявить определяющие успешность игры закономерности

## Предобработка данных

### Названия столбцов

Для начала приведем названия столбцов к нижнему регистру

```python
df.columns = df.columns.str.lower()
#проверим
df.columns
```

### Обработка пропусков и изменение типов данных

#### Название и жанр

Удостоверимся, что по 2 пропуска в названии и жанры - это одни и те же строки

```python
df.query('name.isna()')
```

Да, так и есть, причем обе игры 93 года и на одной платформе. Но это неважно, особенно учитывая отсуствие рейтингов и минимальные объемы продаж для второй. Их можно просто удалить

```python
df = df[(~df['name'].isna())].reset_index(drop=True)
#проверим
df.info()
```

#### Год релиза

Год релиза отсутствует для 269 записей. Это немного, меньше 2%. Вычислить эти данные нeвозможно, а они нам очень пригодятся при анализе. Кроме того, пока есть пропуски, нельз перевести года из float к формату даты, что было бы желательно. Посмотрим, какие у нас вообще есть года может есть какая-то закономерность для этих записей. Может быть, пропуски для одной платформы или одного жанра.

```python
#посмотрим на столбец 'year_of_release'.
print(df['year_of_release'].describe())
#проверим, сколько жанров в играх с пропущенными годами
print('Жанры для игр без года релиза:\n', df.query('year_of_release.isna()')['genre'].value_counts())
#и проверим сколько платформ
print('Платформы для игр без года релиза:\n', df.query('year_of_release.isna()')['platform'].value_counts())
```

Годы от 1980 до 2016. Пропущенные значения есть для разных платформ и жанров. Учитывая небольшой в целом процент, можно просто удалить данные без годов. 

```python
df = df[(~df['year_of_release'].isna())].reset_index(drop=True)
#проверим
df.info()
```

Теперь годы можно вывести из float. Формат datetime нам тут особо не понадобится, только значения годов, поэтому можно просто привести к int

```python
df['year_of_release'] = df['year_of_release'].astype(int)
#проверим
df.info()
```

#### Оценки

Пропуски в столбцах оценок вряд ли можно чем-то заполнить - они зависят главным образом от качества самой игры. Вполне можно представить ситуацию, что игра никак не оценена. Оценки критиков по 100-бальной шкале можно было бы привести к int, если убрать пропуски, но это не имеет практического смысла, а вот оценки пользователей надо перевести в float. Сначала разберемся, какие значения вообще бывают.

```python
df['user_score'].unique()
```

Как, в общем, и ожидалось, значения числовые, кроме одного - 'tbd'. Посмотрим, что это за значение, сколько таких значений. Посмотрим, не зависит ли оно, например, от года, платформы или жанра. Для этого построим столбчатые диаграммы.

```python
print('Всего случаев tbd:', len(df.query('user_score == "tbd"')))
(df.query('user_score == "tbd"').groupby('year_of_release').agg({'name': 'count'}).
 rename(columns = {'name':'total_games_tbd'}).plot(kind = 'bar', grid = True, figsize = (15, 6)))
plt.show()
(df.query('user_score == "tbd"').groupby('platform').agg({'name': 'count'}).
 rename(columns = {'name':'total_games_tbd'}).plot(kind = 'bar', grid = True, figsize = (15, 6)))
plt.show()
(df.query('user_score == "tbd"').groupby('genre').agg({'name': 'count'}).
 rename(columns = {'name':'total_games_tbd'}).plot(kind = 'bar', grid = True, figsize = (15, 6)))
plt.show()
```

Всего случаев tbd 2376. Это очень много - больше 20% от всех известных оценок пользователей.

Никакой конкретной закономерности выделить нельзя, хотя на графике по годам видно, что большая часть игр tbd - до 2012 года. Для нас более актуальными будут более поздние данные - ведь анализ нужно проводить для актуальных платформ. Можно предположить, что с 2012 года как-то изменилась система оценивания, что позволило пользователям легче оценивать игры или легче собирать такие оценки. Простой поисковый [запрос](https://www.google.com/search?q=tbd&oq=tbd&aqs=chrome..69i57j0i433i512j0i512l8.771j0j15&sourceid=chrome&ie=UTF-8) показывает, что это может быть аббревиатура to be determined - требует уточнения. И такое объяснение похоже на правду: чем раньше - тем больше игр с оценками tbd (сложнее было собирать и оценивать - не у всех была возможность просто поставить завездочки на какой-нибудь онлайн-платформе), но с определенного момента назад процент tbd начинает падать, потому что и количество игр было меньше и система оценивания не так сложилась. Для нас, по сути, tbd можно приравнять к NaN. Это, конечно, увеличит число пропусков в и без того не самом полном столбце оценок. Но, во-первых, за последние, самые актуальные годы, 'tbd' не так уже много, а, во-вторых, другого варианта кажется и нет. 

```python
df['user_score'] = df['user_score'].replace('tbd', np.NaN)
#проверим
print('Пропусков должно быть:', len(df) - 9839 + 2376)
print('Фактически:', len(df.query('user_score.isna()')))
```

Теперь можно привести оценки к типу float.

```python
df['user_score'] = df['user_score'].astype(float)
#проверим
df.info()
```

### Платформы

Для начала просто взгялнем на количество игр, выпущенных для разных платформ

```python
df.groupby('platform').agg({'name': 'count'}).rename(columns = {'name':'total'}).plot(kind = 'bar', grid = True, figsize = (15, 6))
plt.show()
```

Количество игр, выпущенных для платформ, очень разное. При это большое количество игр не означает важность для нашего анализа, поскольку платформа может быть давно устаревшей - это касается и самых массовых PS2 и DS. Кроме того, нам важнее одна игра, проданная в 10 млн.копий, чем 10 игр, проданные в 1000 копий каждая. Посмотрим данные по платформам в разрезе количества продаж.

```python
platform_sales = df.groupby('platform').agg({'total_sales': 'sum'})
platform_sales.plot(kind = 'bar', grid = True, yticks=range(0, 1100, 100), figsize = (15, 10))
plt.show()
```

Очевидно, что существуют совсем "карликовые" по количеству продаж платформы и лидеры рынка. Отберем только те, у которых суммарное количество продаж больше 200 млн. и построим для них распределение по годам

```python
#Найдем нужные платформы
top_platforms = platform_sales.query('total_sales > 200').index.tolist()
#построим график
(df.query('platform in @top_platforms').pivot_table(index='year_of_release', columns = 'platform', values='total_sales', aggfunc='sum')
 .plot(grid=True,  style = 'o-', xticks = range(1982, 2016, 1), figsize=(17, 10)))
plt.show()
```

На графике отлично видно, что срок жизни одной платформы составляет, в среднем, 8-10 лет. Есть платформы, существовавшие меньше (вплоть до 5) и чуть больше (11 - 12 лет), но в целом платформа появляется, продажи игр для нее достигают пика через 4-6 лет, а затем снижаются из-за появления новых игровых платформ и устаревания. Единственное исключение, разумеется, PC, но это, по сути и не платформа, а постоянно эволюционирующий отдельный класс.

Из графика можно увидеть платформы, ответственные за определенные пики в продажах по годам. Так, небольшие пики 1998 и 2002 года явно связаны с PS и PS2, а самый большой пик 2008-2009 года  - с Wii, Xbox 360, PSP и, частично, PS3. По сути, можно отследить поколения консолей.

Этот график построен для платформ, продажи для которых уже достигли 200 млн.копий, поэтому в нем, возможно, нет только появляющихся, растущих, самых перспективных платформ, которые и будут, в том числе, нам нужны. Кроме того, надо учитывать, что данные за 2016 год неполные, чем и иожет быть вызвана обманчиво снижающаяся динамика PS4, к примеру. 

Можно сделать вывод, что актуальными для нас будут являться данные за последние 5 лет - половину жизненного цикла платформы, ведь нам нужны те платформы, которые имеют перспективы: еще находятся на пике или растут. Если взять больше, то мы зацепим предыдущие поколения, которые уже совсем сошли на нет, если меньше - то не войдут уже снижающиеся, но еще недавно мощные платформы, типа PS3 и Xbox 360.

Все платформы брать нет необходимости: 
- во-первых, кое где мало оценок, 
- во-вторых, нас не интересует отжившие свое DS и PSP, а также предпоследнее поколение PS и Xbox, когда есть более актуальные.

## Проверка гипотез

Осталось проверить две гипотезы: о равенстве пользовательских рейтингов платформ Xbox One и PC и неравенстве рейтингов жанров Action и Sports

### Рейтинги платформ Xbox One и PC

Нам нужно проверить, гипотезу, что пользовательские рейтинги для платформ Xbox One и PC одинаковые. Для этого будем использовать статистический тест о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. 

Нулевая гипотеза будет звучать так:
- Средние пользовательские рейтинги платформ Xbox One и PC равны

Альтернативная:
- Средние пользовательские рейтинги платформ Xbox One и PC различаются

Поскольку количество игр, т.е. размер совокупностей, для каждой платформы разный, установим equal_var = False. Порог статистической значимости установим 5%. Кроме того, у нас довольно много пропусков в столбце рейтинга пользователей - надо убрать их из выборки

```python
alpha = .05
results_xb_pc = st.ttest_ind(df_actual.query('platform == "XOne" and user_score.notna()')['user_score'], 
                             df_actual.query('platform == "PC" and user_score.notna()')['user_score'], 
                             equal_var = False)

print('p-значение: ', results_xb_pc.pvalue)

if results_xb_pc.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Имеющиеся данные указывают на то, гипотеза о равенстве средних пользовательских рейтинги платформ Xbox One и PC может быть верна.

### Рейтинги жанров Action и Sports

Теперь проверим гипотезу, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные. Нужно проверить гипотезу о разности, но нулевая все равно должна быть о равенстве:
- Средние пользовательские рейтинги жанров Action и Sports равны

Альтернативная гипотеза (ее нам на самом деле и надо проверить):
- Средние пользовательские рейтинги жанров Action и Sports различаются

Как и в предыдущем случае есть основания полгать что размер совокупностей, как и дисперсия, разный, поэтому надо установить equal_var = False. Порог значимости оставим тем же. И уберем игры без оценки

```python
results_action_sports = st.ttest_ind(df_actual.query('genre == "Action" and user_score.notna()')['user_score'], 
                             df_actual.query('genre == "Sports" and user_score.notna()')['user_score'], 
                             equal_var = False)

print('p-значение: ', results_action_sports.pvalue)

if results_action_sports.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

P-значение чрезвычайно мало, следовательно можно отвергнуть нулевую гипотезу о равенстве рейтингов и принять альтернативную:
- Средние пользовательские рейтинги жанров Action и Sports разные
