## Предобработка данных

### Обработка пропусков

#### Балконы

Для начала изучим имеющиеся в данных пропуски. Начнем с балконов, как самого простого

```python
#посмотрим данные о балконах
#какие значения бывают
print(df['balcony'].value_counts())
#сколько пропусков
df['balcony'].isna().sum()
```

Не указаны, скорее всего, те балконы, которых нет. Заполним пропуски нулями

```python

df['balcony'] = df['balcony'].fillna(0)
#проверяем
df['balcony'].isna().sum()
```

#### Высота потолков

В столбце ceiling_height почти половина - пропуски. Очевидно, люди не всегда указывают высоту потолков в объявлении. Можно было бы просто заполнить медианным значением. Для начала посмотрим на разброс значений вообще

```python
df['ceiling_height'].describe()
```

<a id='1'></a> Очевидно, есть выбросы - ошибки в данных, причем как в меньшую, так и в большую сторону. Посмотрим, сколько и какие значения у строк, где потолки выше 10 м и меньше 2. 10 м - тоже маловероятный вариант, но какие-то лофтовые помещения могут быть. 

```python
df.query('ceiling_height > 10 or ceiling_height < 2')['ceiling_height'].value_counts()
```

Кажется вероятным, что все числа больше 20 - это просто опечатка - не там поставленный разделитель. Надо поделить на 10. Что делать со странными потолками до 2 м - неясно. С одной стороны это может быть опечатка того же типа - не там поставили разделитель, но потолок 17.5 - не менее странен, чем 1.75, а значения - единичные. Оставим как есть

```python
df.loc[df['ceiling_height'] > 20,'ceiling_height'] = df.loc[df['ceiling_height'] > 20, 'ceiling_height'] / 10
#проверяем
df.query('ceiling_height > 10 or ceiling_height < 2')['ceiling_height'].value_counts()
```

Высота потолков пригодится нам в анализе, а пропусков - почти половина. Желательно бы их заполнить. Можно заполнить просто медианой, а можно - медианой, сгруппированной по поселениям. Логика в том, что разные места застраивались в разное время, и, соотвественно, в них превалируют разные типы зданий и высота потолков. Для старого и большого Питера, с самой различной застройкой это не так показательно, а для условного Мурино с огромный количеством однотипного жилья - вполне. Посмотрим на столбец 'locality_name'.

```python
df['locality_name'].value_counts()
```

Похоже, в данных название смешано с типом поселения, который еще и может быть по-разному написан. Избавимся от лишнего. Посмотрим какие типы вообще бывают и надо убрать. Применим лемматизацию, чтоб выявить все типы

```python
from pymystem3 import Mystem
m = Mystem()
from collections import Counter

locnam = df[df['locality_name'].notna()]
locnam = locnam['locality_name'].unique()
locnam = ' '.join(locnam)
locnam = m.lemmatize(locnam)
Counter(locnam)
```

В списке явно есть:
- поселок;
- поселок городского типа
- деревня
- садовое товарищество
- село тоже есть, но это может быть часть названия, лучше оставить
- коттеджный поселок
- садоводческое некоммерческое товарищество

И все то же с буквой ё в слове поселок. Уберем лишнее

```python
names = (['поселок городского типа ', 
          'посёлок городского типа ', 
          'коттеджный поселок ', 
          'коттеджный посёлок ', 
          'поселок ', 
          'посёлок ', 
          'деревня ', 
          'садоводческое некоммерческое товарищество ', 
          'садовое товарищество '])

for name in names:
         df['locality_name'] = df['locality_name'].str.replace(name, '')
         
#проверим
df['locality_name'].value_counts()
```

Осталось много одиночных странных названий, но крупные группы объединились (то же Мурино - 590 строк вместо 522, Кудрово внезапно возникло на 3 месте и т.д.). Осталось понять, что делать с 49 пустыми значениями. Я думаю, что это ничтожно мало, а местоположение - ключевое значение для объекта недвижимости. Поэтому просто исключим их из данных

```python
df = df[(~df['locality_name'].isna())].reset_index(drop=True)
#проверим
df.info()
```

А теперь вернемся к потолкам. Создадим столбец, в котором были бы данные по медианной высоте потолка для поселений и заполним его значениями пустые ячейки

```python
#создадим фрейм с группированными значениями
group_med_ceil = (df.groupby(['locality_name']).agg({'ceiling_height': 'median'}).
                  rename(columns = {'ceiling_height': 'median_ceiling_height'}))
#дополняем исходный фрейм этими данными
df = df.merge(group_med_ceil, on = ['locality_name'])
#переносим в пустые ячейки 'total_income' значения 'median_total_income'
df.loc[df['ceiling_height'].isna(), 'ceiling_height'] = df.loc[df['ceiling_height'].isna(), 'median_ceiling_height']
#проверяем пропуски
df['ceiling_height'].isna().sum()
```

Остались 102 значения, для которых не нашлось данных в группировке по населенным пунктам. Заполним их просто медианой

```python
df.loc[df['ceiling_height'].isna(), 'ceiling_height'] = df['ceiling_height'].median()
#проверяем пропуски
df['ceiling_height'].isna().sum()
```

#### Жилая площадь и площадь кухни

Эти показатели по своей природе похожи, и логично обработать их вместе. Но сначала посмотрим на кухню.
    Некоторые пропуски могут быть связаны с особенностями квартиры. В частности kitchen_area может отсутствовать, потому что квартира является студией или свободной планировки. Проверим, для этого посмотрим, сколько квартир имеют свободную планировку или являются студиями и одновременно для них не указана площадь кухни 

```python
#Посмотрим, сколько всего объявлений со свободной планировкой, и в скольки из них не указана кухня
print('Свободная планировка:', (len(df.query('open_plan == True'))))
print('Свободная планировка, не указана кухня:', (len(df.query('open_plan == True and kitchen_area.isna()'))))
#То же для студий
print('Студий всего:', (len(df.query('studio == True'))))
print('Студия, не указана кухня:', (len(df.query('studio == True and kitchen_area.isna()'))))
#Не свободной планировки и не студии, но площадь не указана:
print('Не указана кухня, не свободной планировки, не студия:', 
      (len(df.query('open_plan == False and studio == False and kitchen_area.isna()'))))
```

Действительно, для всех студий и квартир свободной планировки закономерно не указана площадь кухни, но доля таких квартир невелика. Для них логично будет установить площадь кухни равной нулю

```python
df.loc[df['studio'] == True, 'kitchen_area'] = 0
df.loc[df['open_plan'] == True, 'kitchen_area'] = 0
#проверяем
len(df.query('open_plan == False and studio == False and kitchen_area.isna()'))
```

Одну особенность кухонь учли, осталось 2009 значений. Теперь посмотрим на жилую площадь. Можно заполнить пропуски на основании данных об общей площади и количестве комнат.

```python
#Посмотрим на значения
print('Пропусков в living_area:', df['living_area'].isna().sum())
df['living_area'].describe()
```

И жилую площадь и площадь кухни можно заполнить на основании данных об общей площади и количестве комнат, рассчитав соответствующие соотношения жилой и кухонной площади по отношению к общей для квартир с разным количеством комнат.

```python
#найдем коэффициенты отношения жилой и кухонной площади к общей для квартир с разным количеством комнат.
#Для это сгруппируем данные по количеству комнат, а потом добавим соотвествующие столбцы
ratio_data = (df.groupby(['rooms']).agg({'total_area': 'median', 'living_area': 'median', 'kitchen_area': 'median'}).
                     rename(columns = {'total_area': 'median_total_area', 'living_area': 'median_living_area', 
                                       'kitchen_area': 'median_kitchen_area'}))
ratio_data['ratio_living'] = ratio_data['median_living_area']/ratio_data['median_total_area']
ratio_data['ratio_kitchen'] = ratio_data['median_kitchen_area']/ratio_data['median_total_area']
#добавляем данные в основную таблицу
df = df.merge(ratio_data, on = ['rooms'])
#заменяем значения в пустых ячейках living_area, умножая значение total_area на коэффициент
df.loc[df['living_area'].isna(), 'living_area'] = (df.loc[df['living_area'].isna(), 'total_area'] * 
                                                   df.loc[df['living_area'].isna(), 'ratio_living'])
df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = (df.loc[df['kitchen_area'].isna(), 'total_area'] * 
                                                   df.loc[df['kitchen_area'].isna(), 'ratio_kitchen'])
#проверяем
print('Пропусков в living_area:', df['living_area'].isna().sum())
print('Пропусков в kitchen_area:', df['kitchen_area'].isna().sum())
df.head()
```
